SVD 모델은 사용자- 아이템 - 평점 행렬을 가지고와서 잘게 분해해서 그안에 숨겨진 취향을 찾아서 추런해주는 모델

|-------|----|----|-|----|
|유저1|4.0|null|3.0|...|
|유저101|3.5|4.0|null|...|

사용자 - 아이템 행렬

```python
# 잠재요인 협업필터링 추천시스템을 함수로 만든다
# 유저에게 추천 -> 1) 유저가 안본 영화를 찾는다. 2) 안본 영화의 평점을 찾는다. 3) 정렬해서 보여준다
movies = pd.read_csv('./data/movies.csv')
movies
```

```python
# 유저가 안본 영화를 반환하는 함수
def get_unseen_surprise(reatings,movies,userId):
    seen_movies = df[df['userId']==userId]['movieId'].tolist()  # 유저가 본 영화 list
    total_movies = movies['movieId'].tolist()                   # 전체 영화 list

    unseen_movies = []                  # 안본 영화 리스트
    for movie in total_movies:          # 전체 영화 리스트를 순회
        if movie not in seen_movies:    # 본 영화가 아니면 
            unseen_movies.append(movie) # 안본 영화 리스트에 추가

    return unseen_movies
```

```python
unseen_movies = get_unseen_surprise(df,movies,9)
len(unseen_movies)
unseen_movies
```

```python
# SVD 모델을 이용하여 특정 사용자가 아직 보지 않은 영화중에서 좋아할 학률이 가장 놓은 top-N개의 영화를 추천해주는 함수
def recommend_movie_by_surprise(model,userid,unseen_movies,top_n=5):
    pred_list = []
    for movieid in unseen_movies:
        pred = model.predict(str(userid),str(movieid))  # 결과는 Prediction(uid='9', iid='41', r_ui=None, est=3.3731093250759443, details={'was_impossible': False})
        pred_list.append(pred)

    def sort_key(pred):
        return pred.est # 예측 평점(est)

    pred_list.sort(key=sort_key,reverse=True)  # 예측 쳥점 기준으로 내림차순 정렬

    top_pred =pred_list[:top_n] # 상위 n개의 추천 결과

    result = []
    for temp in top_pred:
        dict_temp = {f"영화id:{temp.iid}, \
                     rating:{temp.est},\
                    영화명: {movies[movies['movieId'] == int(temp.iid)]['title']}"}
        result.append(dict_temp)

    return result
        
    
```

```python
recommend_movie_by_surprise(model, 9, unseen_movies, top_n=10)
# unseen_movies
```
