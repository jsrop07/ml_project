```python
retail_df = retail_df.dropna()  #결측치 제거

# IQR 방식으로 이상치 제거하는 함수
def remove_outlier_iqr(df,column):
    Q1 = df[column].quantile(0.25)  # 1사분위수
    Q3 = df[column].quantile(0.75)  # 3사분위수
    
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return df[(df[column]>=lower) & (df[column]<= upper) & (df[column] > 0)]

# Quantity, UnitPrice 이상치 제거
retail_df =  remove_outlier_iqr(retail_df,'UnitPrice')   #retail_df.dropna(subset=['Quantity','UnitPrice'])
retail_df =  remove_outlier_iqr(retail_df,'Quantity')
retail_df
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.boxplot(retail_df['Quantity'])
plt.title('IQR 적용 후 Quantity')

plt.subplot(1,2,2)
plt.boxplot(retail_df['UnitPrice'])
plt.title('IQR 적용 후 UnitPrice')

plt.show()
```

```python
# 수량과 단가를 곱해서 거래금액 컬럼 생성, 고객 ID는 정수형으로 변환
retail_df['sale_amount'] =  retail_df['Quantity'] * retail_df['UnitPrice']
retail_df['CustomerID'] = retail_df['CustomerID'].astype(int)

retail_df
```

```python
# 어떤 고객이(고객ID로 그룹핑) 가장 많이 샀을까? (sale_amount를 sum해주고, 그 순위를 내림차순으로 정렬)

retail_df.groupby('CustomerID')['sale_amount'].sum().sort_values(ascending=False)
```

```python
# 가장 많이 구매한 고객 상위 10명
top10 = (retail_df.groupby('CustomerID')['sale_amount'].sum().sort_values(ascending=False).head(10).sort_values())
plt.figure(figsize=(8,5))
top10.plot(kind = 'barh')
plt.title(' 총 구매 금액 상위 10명 고객')
plt.xlabel(' 구매 고객')
plt.ylabel('고객')
plt.show()
```

```python
# 상위 10% 고객이 차지하는 매출 비중 계산
customer_sales = retail_df.groupby('CustomerID')['sale_amount'].sum().sort_values(ascending=False)

total_customers = customer_sales.shape[0]
top10_percent_cnt = int(total_customers * 0.1)
top10_percent_cnt
top10_sales= customer_sales.iloc[:top10_percent_cnt].sum()  # 상위 10% 고객 매출
top10_sales 
total_sales = customer_sales.sum()
top10_ratio = (top10_sales/total_sales)*100
print(f"상위 10% 고객이 차지하는 매출 비중 : {top10_ratio:.1f}")
```

상위 10% 고객이 매출의 거의 절반을 차지할만큼 소수 고객에세 매출이 집중되어있다

```python
# 고객 id를 기준으로 각종 데이터를 집계
aggregation ={
    'sale_amount' : 'sum',  # 고객별 총 구매금액
    'InvoiceNo' : 'count',  # 고객별 구매 횟수
    'InvoiceDate' : 'max',  # 고객의 마지막 구매 일자
}

cust_df = retail_df.groupby('CustomerID').agg(aggregation)  # 고객 단위로 각종 지표 집계
cust_df
```

```python
# 기준일자를 설정하고, 기준일자(데이터의 마지막일자 + 1) 기준 최근 구매일로부터 경과한 일수를 숫자 데이터로 변환해 계산
import datetime as dt
cust_df['InvoiceDate'] = dt.datetime(2011,12,10) - cust_df['InvoiceDate'] # 일자로 반환
cust_df
```

```python
# 최근 구매 경과시간을 일 단위로 변환하여 인공지능 모델리에 사용 가능하게 만듦
cust_df['InvoiceDate'] = cust_df['InvoiceDate'].dt.days # 일시 -> 일자(day)로 변경
cust_df
```

```python
# 스케일링 및 군집 분석
from sklearn.preprocessing import StandardScaler    # 평균 0, 표준편차 1로 스케일링 해주는 도구
from sklearn.cluster import KMeans                  # 군집화 모델(비지도 학습)
from sklearn.metrics import silhouette_samples, silhouette_score    # 군집 품질 평가 도구
```

```python
scalerd_data = StandardScaler().fit_transform(cust_df)
scalerd_data
```

```python
kmeans =KMeans(n_clusters=2, random_state=0)
kmeans.fit(scalerd_data)    # 군집 학습
kmeans.labels_              # 군집 할당 결과
```

```python
# cust_df의 cluster파생변수를 만들어 군집 할당 결과를 저장
cust_df['cluster'] = kmeans.labels_ 

cust_df
```

```python
silhouette_score(
    scalerd_data,   # 스케일링 입력 데이터
    kmeans.labels_  # 군집 라벨
    )
```
